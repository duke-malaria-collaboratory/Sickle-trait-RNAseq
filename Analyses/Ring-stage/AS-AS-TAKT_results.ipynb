```python
# just Ring stage genes (peak expression at or before 24 hours)
AS_AS_TAKT_results = pd.read_csv("Results/AS-AS-Results/kt_diss_AS_3D7_1stage_periodic_jtk05-AS_FUP_1stage_periodic_jtk05.tsv", 
                                     sep="\t", comment="#", index_col=0)
```

```python
time_points = list(range(3, 51,3))
# AA_3D7_1stage_periodic_ts_df = pd.read_csv("Data/AA_3D7_1stage_periodic_jtk05.txt", sep="\t", index_col=0)
# AA_3D7_1stage_periodic_ts_df.columns = time_points
AS_3D7_1stage_periodic_ts_df = pd.read_csv("Data/AS_3D7_1stage_periodic_jtk05.txt", sep="\t", index_col=0)
AS_3D7_1stage_periodic_ts_df.columns = time_points
# AA_FUP_1stage_periodic_ts_df = pd.read_csv("Data/AA_FUP_1stage_periodic_jtk05.txt", sep="\t", index_col=0)
# AA_FUP_1stage_periodic_ts_df.columns = time_points
AS_FUP_1stage_periodic_ts_df = pd.read_csv("Data/AS_FUP_1stage_periodic_jtk05.txt", sep="\t", index_col=0)
AS_FUP_1stage_periodic_ts_df.columns = time_points
```

```python
AS_AS_shifted_genes = AS_AS_TAKT_results[(AS_AS_TAKT_results["est_delay"] > 0.0) & (AS_AS_TAKT_results["kt_diss_pval"] <= 0.05)]
```

```python
AS_AS_shifted_genes.sort_values(by=['est_delay'])
```

```python
gene_name = "PF3D7_1122800"
gene_df = pd.concat([AS_3D7_1stage_periodic_ts_df[AS_3D7_1stage_periodic_ts_df.index == gene_name],
                             AS_FUP_1stage_periodic_ts_df[AS_FUP_1stage_periodic_ts_df.index == gene_name]])
gene_df.index = [gene_name + "_AS_3D7", gene_name + "_AS_FUP"]
gene_df.reset_index(inplace=True)
gene_df.rename(columns={"index": "gene"}, inplace=True)
gene_df = pd.melt(gene_df, id_vars="gene", value_name="TPM", var_name="Hours")
sns.lineplot(x="Hours", y="TPM", hue="gene", data=gene_df)
```
