```python
def prep_df_for_plot(takt_df):
    new_takt_df = takt_df.reset_index()
    datasets = new_takt_df.columns[0]
    new_takt_df = new_takt_df[[datasets, 'kt_diss', 'kt_diss_pval', 'est_delay']]
    new_takt_df['datasets'] = datasets
    new_takt_df = new_takt_df.rename(columns={datasets: "gene"})
    return(new_takt_df)

def correct_est_delay(takt_df):
    corrected_takt_df = takt_df.copy()
    for index, row in corrected_takt_df.iterrows():
        est_delay = row[3]
        new_est_delay = 0
        if est_delay > 24.0:
            new_est_delay = est_delay - 24.0
            corrected_takt_df.at[index, 'est_delay'] = new_est_delay
    return(corrected_takt_df)
```

```python
# 3D7 col drops for melting and plotting
sub_AA13_3D7_AA17_3D7_takt = correct_est_delay(prep_df_for_plot(AA13_3D7_AA17_3D7_takt))
sub_AA13_3D7_AS15_3D7_takt = correct_est_delay(prep_df_for_plot(AA13_3D7_AS15_3D7_takt))
sub_AA13_3D7_AS16_3D7_takt = correct_est_delay(prep_df_for_plot(AA13_3D7_AS16_3D7_takt))
sub_AA17_3D7_AS15_3D7_takt = correct_est_delay(prep_df_for_plot(AA17_3D7_AS15_3D7_takt))
sub_AA17_3D7_AS16_3D7_takt = correct_est_delay(prep_df_for_plot(AA17_3D7_AS16_3D7_takt))
```

```python
# FUP col drops for melting and plotting
sub_AA13_FUP_AA17_FUP_takt = correct_est_delay(prep_df_for_plot(AA13_FUP_AA17_FUP_takt))
sub_AA13_FUP_AS18_FUP_takt = correct_est_delay(prep_df_for_plot(AA13_FUP_AS18_FUP_takt))
sub_AA13_FUP_AS19_FUP_takt = correct_est_delay(prep_df_for_plot(AA13_FUP_AS19_FUP_takt))
sub_AA17_FUP_AS18_FUP_takt = correct_est_delay(prep_df_for_plot(AA17_FUP_AS18_FUP_takt))
sub_AA17_FUP_AS19_FUP_takt = correct_est_delay(prep_df_for_plot(AA17_FUP_AS19_FUP_takt))
```
